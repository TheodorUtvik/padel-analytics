## Matches

```python
import datetime

# Fetch completed matches only — today's date as upper bound
today = datetime.date.today().isoformat()
matches = client.get_matches(before_date=today)
df_matches = pd.DataFrame(matches) if isinstance(matches, list) else pd.DataFrame(matches.get('data', matches))
print(df_matches.shape)
df_matches.head()
```

## Head-to-Head

```python
# Use real players from match data — team_1 vs team_2 from the first match
first_match = df_matches.iloc[0]['players']
team_1_ids = [p['id'] for p in first_match['team_1']]
team_2_ids = [p['id'] for p in first_match['team_2']]

print(f'team_1: {team_1_ids}, team_2: {team_2_ids}')

h2h = client.get_head_to_head(team_1=team_1_ids, team_2=team_2_ids)
df_h2h = pd.DataFrame(h2h.get('data', []))
print(df_h2h.shape)
df_h2h.head()
```

## Pair Matches

```python
# Extract real pairs from match data — players must have actually played together
def extract_pairs(matches_data):
    pairs = []
    for match in matches_data:
        players = match.get('players', {})
        for team_key in ('team_1', 'team_2'):
            team = players.get(team_key, [])
            if len(team) == 2:
                pairs.append((team[0]['id'], team[1]['id']))
    return pairs

pairs = extract_pairs(df_matches.to_dict('records'))
print(f'Found {len(pairs)} pairs from match data')
print('First few:', pairs[:5])

if pairs:
    p1_id, p2_id = pairs[0]
    print(f'\nFetching matches for pair: {p1_id}-{p2_id}')
    pair_matches = client.get_pair_matches(p1_id, p2_id)
    df_pair_matches = pd.DataFrame(pair_matches.get('data', []))
    print(df_pair_matches.shape)
    display(df_pair_matches.head())
else:
    print('No pairs found — try fetching matches with a broader date range')
```

## Notes

### Free tier endpoints (available ✅)
- `GET /api/seasons` + `/api/seasons/{id}/tournaments`
- `GET /api/players` + `/api/players/{id}` + `/api/players/{id}/matches`
- `GET /api/matches` + `/api/matches/{id}`
- `POST /api/matches/head-to-head`
- `GET /api/tournaments` + `/api/tournaments/{id}` + `/api/tournaments/{id}/matches`
- `GET /api/pairs/{p1}-{p2}` + `/api/pairs/{p1}-{p2}/matches`

### Locked endpoints (require paid plan 🔒)
- `GET /api/players/{id}/stats`
- `GET /api/matches/{id}/stats` (point-by-point)
- `GET /api/pairs/{p1}-{p2}/stats`
- `GET /api/live/matches`
