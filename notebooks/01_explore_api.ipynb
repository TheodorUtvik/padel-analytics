## Notes

### Free tier endpoints (available ✅)
- `GET /api/seasons` + `/api/seasons/{id}/tournaments`
- `GET /api/players` + `/api/players/{id}` + `/api/players/{id}/matches`
- `GET /api/matches` + `/api/matches/{id}`
- `POST /api/matches/head-to-head`
- `GET /api/tournaments` + `/api/tournaments/{id}` + `/api/tournaments/{id}/matches`
- `GET /api/pairs/{p1}-{p2}` + `/api/pairs/{p1}-{p2}/matches`

### Locked endpoints (require paid plan 🔒)
- `GET /api/players/{id}/stats`
- `GET /api/matches/{id}/stats` (point-by-point)
- `GET /api/pairs/{p1}-{p2}/stats`
- `GET /api/live/matches`

### Project direction (free tier)
- **Match outcome prediction** — from win/loss results + round reached per tournament
- **Player form analysis** — rolling win rate over recent matches
- **Pair chemistry** — win %, round reached, opponents faced together
- **ELO / ranking model** — iterative rating from raw match results
